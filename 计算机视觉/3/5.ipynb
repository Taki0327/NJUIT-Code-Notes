```python
from skimage import data
import matplotlib.pyplot as plt
import numpy as np
img = data.chelsea()
test = np.zeros_like(img, dtype=np.uint8)
img_gray= np.zeros((img.shape[0], img.shape[1]),dtype=np.uint8)
for i in range(img.shape[0]):
    for j in range(img.shape[1]):
        img_gray[i,j] = np.dot(img[i,j,:], (0.3,0.59,0.11))
#
L=255#
def get_r(intensity):
    if intensity<L/2:
        _update=0
    elif intensity>3*L/4:
        _update=L
    else:
        _update=4*intensity-2*L
    return _update
def get_g(intensity):
    if intensity<L/4:
        _update=4*intensity
    elif intensity>3*L/4:
        _update=4*L-4*intensity
    else:
        _update=L
    return _update
def get_b(intensity):
    if intensity<L/4:
        _update=L
    elif intensity>L/2:
        _update=0
    else:
        _update=2*L-4*intensity
    return _update
for i in range(img_gray.shape[0]):
    for j in range(img_gray.shape[1]):
        tmp_intensity=img_gray[i,j]
        test[i,j]=(get_r(tmp_intensity),get_g(tmp_intensity),get_b(tmp_intensity))

fig = plt.figure(figsize=(21,9))
org = fig.add_subplot(1,3,1)
org.imshow(img)
org.set_title("orginal image")
_img = fig.add_subplot(1,3,2)
_img.imshow(img_gray,cmap='gray')
_img.set_title("_img image")
_test = fig.add_subplot(1,3,3)
_test.imshow(test,cmap='gray')
_test.set_title("test image")
plt.tight_layout()
plt.show()
```
